```python
#Coding 6
# Mengvisualisasi pada data yang telah diberikan
# Grafik
import pandas as pd
import matplotlib.pyplot as plt

# Memuat data dari URL
df = pd.read_csv('https://raw.githubusercontent.com/Alam-A99/Pandas_Latih/main/penjualan.csv')

# Menghitung total penjualan untuk bulan Maret dan April
total_maret = df['mar'].sum()
total_april = df['apr'].sum()
total_feb = df['feb'].sum()

# Membuat DataFrame baru untuk plotting
data_penjualan = pd.DataFrame({
    'Bulan': ['Maret', 'April','feb'],
    'Total Penjualan': [total_maret, total_april, total_feb]
})

# Membuat grafik batang dengan Matplotlib
plt.figure(figsize=(8, 6))
plt.bar(data_penjualan['Bulan'], data_penjualan['Total Penjualan'], color=['#34A853', '#4285F4', '#00FFFF'])

# Menambahkan label dan judul
plt.title('Grafik Total Penjualan Bulan Maret, April dan maret', fontsize=16)
plt.xlabel('Bulan', fontsize=12)
plt.ylabel('Total Penjualan', fontsize=12)
plt.xticks(rotation=0)

# Menambahkan nilai di atas setiap bar
for index, value in enumerate(data_penjualan['Total Penjualan']):
    plt.text(index, value, str(value), ha='center', va='bottom', fontsize=10)

# Menyimpan grafik
plt.savefig('grafik_penjualan.png')

print("Grafik berhasil dibuat dan disimpan sebagai 'grafik_penjualan.png'")
```

```python
#Coding 7
# Mengvisualisasi data yang telah diberikan ke grafik pie
import pandas as pd
import matplotlib.pyplot as plt

# Memuat data dari URL
df = pd.read_csv('https://raw.githubusercontent.com/Alam-A99/Pandas_Latih/main/penjualan.csv')

# Menghitung total penjualan untuk bulan Maret dan April
total_maret = df['mar'].sum()
total_april = df['apr'].sum()
total_feb = df['feb'].sum()

# Membuat daftar data dan label
data_penjualan = [total_maret, total_april, total_feb]
label_bulan = ['Maret', 'April', 'Februari']

# Membuat grafik lingkaran
plt.figure(figsize=(8, 8))
plt.pie(data_penjualan, labels=label_bulan, autopct='%1.1f%%', startangle=90, colors=['#34A853', '#4285F4', '#00FFFF'])

# Menambahkan judul
plt.title('Total Penjualan Bulan Maret, April dan Februari', fontsize=16)

# Menyimpan grafik
plt.savefig('grafik_penjualan_pie.png')

print("Grafik lingkaran berhasil dibuat dan disimpan sebagai 'grafik_penjualan_pie.png'")
```

```python
#Coding 9
# Mengvisualisasi data tersebut menggunakan grafik line
import pandas as pd
import matplotlib.pyplot as plt

# Memuat data dari URL
df = pd.read_csv('https://raw.githubusercontent.com/Alam-A99/Pandas_Latih/main/penjualan.csv')

# Menghitung total penjualan untuk bulan Maret dan April
total_maret = df['mar'].sum()
total_april = df['apr'].sum()

# Membuat daftar data dan label untuk grafik garis
bulan = ['Maret', 'April']
penjualan = [total_maret, total_april]

# Membuat grafik garis
plt.figure(figsize=(8, 6))
plt.plot(bulan, penjualan, marker='o', linestyle='-', color='red')

# Menambahkan label dan judul
plt.title('Tren Penjualan Bulan Maret dan April', fontsize=16)
plt.xlabel('Bulan', fontsize=12)
plt.ylabel('Total Penjualan', fontsize=12)
plt.grid(True)

# Menyimpan grafik
plt.savefig('grafik_penjualan_line.png')

print("Grafik garis berhasil dibuat dan disimpan sebagai 'grafik_penjualan_line.png'")
```

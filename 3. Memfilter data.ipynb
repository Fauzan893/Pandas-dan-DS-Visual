```python
#Coding 7
# Membuat DataFrame dengan beberapa nilai yang hilang
# Modifikasi yang tidak ada dimodul

import pandas as pd
import numpy as np

data = {'Kolom_A': [1, 2, np.nan, 4],
        'Kolom_B': [5, np.nan, 7, 8]}
df = pd.DataFrame(data)

print("DataFrame Asli:")
print(df)

# Mengisi semua nilai yang hilang dengan 0
df_filled = df.fillna(0)
print("\nDataFrame setelah diisi dengan 0:")
print(df_filled)

# Mengisi nilai yang hilang di Kolom_B dengan nilai rata-rata dari kolom tersebut
df_mean = df.fillna(df.mean())
print("\nDataFrame setelah diisi dengan nilai rata-rata:")
print(df_mean)
```

```python
#Coding 8
# Menghapus Baris atau Kolom
# Memodifikasi yang tidak ada dimodul
data = {'Nama': ['Budi', 'Ani', 'Caca'],
        'Usia': [25, 30, 22],
        'Kota': ['Jakarta', 'Bandung', 'Surabaya']}
df = pd.DataFrame(data)

print("DataFrame Asli:")
print(df)

# Menghapus kolom 'Kota'
df_no_city = df.drop(columns=['Kota'])
print("\nDataFrame setelah menghapus kolom 'Kota':")
print(df_no_city)

# Menghapus baris dengan indeks 1 (Ani)
df_no_ani = df.drop(index=1)
print("\nDataFrame setelah menghapus baris dengan indeks 1:")
print(df_no_ani)
```

```python
#Coding 9
# Memodifikasi yang tidak ada di modul
# Mengkelompokkan Data
data = {'Produk': ['Laptop', 'Mouse', 'Laptop', 'Keyboard', 'Mouse'],
        'Region': ['Utara', 'Selatan', 'Utara', 'Timur', 'Selatan'],
        'Penjualan': [1000, 50, 1500, 200, 75]}
df = pd.DataFrame(data)

print("DataFrame Asli:")
print(df)

# Mengelompokkan data berdasarkan 'Produk' dan menjumlahkan 'Penjualan'
penjualan_per_produk = df.groupby('Produk')['Penjualan'].sum()
print("\nTotal Penjualan per Produk:")
print(penjualan_per_produk)

# Mengelompokkan berdasarkan 'Region' dan 'Produk'
penjualan_per_region_produk = df.groupby(['Region', 'Produk'])['Penjualan'].sum()
print("\nTotal Penjualan per Region dan Produk:")
print(penjualan_per_region_produk)
```

```python
#Coding 10
# Memodifikasi yang tidak ada dimodul
# Menghapus baris duplikat
data = {'Nama': ['Andi', 'Budi', 'Andi', 'Caca', 'Budi'],
        'Usia': [25, 30, 25, 22, 30],
        'Kota': ['Jakarta', 'Bandung', 'Jakarta', 'Surabaya', 'Bandung']}
df = pd.DataFrame(data)

print("DataFrame Asli:")
print(df)

# Hapus semua baris duplikat
df_no_dups_all = df.drop_duplicates()
print("\nDataFrame setelah menghapus semua baris duplikat:")
print(df_no_dups_all)

# Hapus baris duplikat berdasarkan kolom 'Nama' saja
df_no_dups_name = df.drop_duplicates(subset=['Nama'])
print("\nDataFrame setelah menghapus duplikat berdasarkan 'Nama':")
print(df_no_dups_name)
```

```python
#Coding 11
# Memodifikasi yang tidak ada dimodul
# Mengubah tipe data
data = {'Kolom_A': [1.5, 2.0, 3.8],
        'Kolom_B': [4, 5, 6]}
df = pd.DataFrame(data)

print("DataFrame Asli:")
print(df)
print("\nTipe data asli:")
print(df.dtypes)

# Ubah tipe data 'Kolom_A' dari float menjadi integer
df['Kolom_A'] = df['Kolom_A'].astype(int)

print("\nDataFrame setelah mengubah 'Kolom_A' ke integer:")
print(df)
print("\nTipe data setelah diubah:")
print(df.dtypes)
```
